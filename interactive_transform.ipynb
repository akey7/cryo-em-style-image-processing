```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def fourier_transform_2D(data):
    # Compute the 2D Fourier transform of the data
    f_transform = np.fft.fft2(data)

    # Shift the zero-frequency component to the center
    f_shift = np.fft.fftshift(f_transform)

    # Compute magnitude spectrum
    magnitude_spectrum = np.abs(f_shift)

    return magnitude_spectrum
```

```python
img = Image.open("input/sines_h=1_k=1.png").convert("L")
spectrum = fourier_transform_2D(img)
```

```python
plt.imshow(np.log(1 + spectrum), cmap='gray')
plt.title('Magnitude Spectrum')
plt.colorbar()
plt.savefig(os.path.join("output", "mag1.png"))
plt.show()
```
